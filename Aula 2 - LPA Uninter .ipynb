#Manipulações avançadas com strings

###Concantenação

```python
s1 = "Lógica de Programação"
s1 = s1 + 'e Algoritmos'
print(s1)
```

Repetindo strings na concatenção:

```python
s1 = "A" + "-" * 10 + "B"
print(s1)
```

```python
s1 = "A" + " " * 10 + "B"
print(s1)
```

#Composição por marcador de posição

```python
nota = 8.5
s1 = "Você tirou %f na disciplina de Algoritmos" % nota
print(s1)
```

Limitando as casas decimais:

```python
nota = 8.5
s1 = "Você tirou %.2f na disciplina de Algoritmos" % nota
print(s1)
```

Várias variáveis:

```python
nota = 8.5
disciplina = "Algoritmos"
s1 = "Você tirou %.2f na disciplina de %s" % (nota,disciplina)
print(s1)
```

#Composição moderna

```python
nota = 8.5
disciplina = "Algoritmos"
s1 = "Você tirou {} na disciplina de {}" .format(nota,disciplina)
print(s1)
```

#Composição com f-string

```python
nota = 8.5
disciplina = "Algoritmos"
s1 = f"Você tirou {nota} na disciplina de {disciplina}"
print(s1)
```

##Convertendo dados de entrada (casting)

```python
nota = float(input("Qual nota voce recebeu na disciplina?"))
print(f"Voce tirou nota {nota}.")
```

##Fluxo de Execução do programa (teste de mesa)

```python
x = 1
y = 1
z = x + y

x = x + 2
y = y - 1
z = x + y

x = y + 1
y = x - 1
z = x + y

print(z)
```
