Exercício 1

Faça um algoritmo que receba três valores, representando os lados de um triângulo fornecidos pelo usuário. Verifique se os valores formam um triangulo e classifique como:

a)Equiláterio (três lados iguais)

b)Isósceles (dois lados iguais)

c)Escaleno (três lados diferentes)

```python
A = int(input("Digite o primeiro lado: "))
B = int(input("Digite o segundo lado: "))
C = int(input("Digite o terceiro lado: "))

if(((A>0) and (B>0) and (C>0)) and ((A+B>C) and (A+C>B) and (B+C>A))):
  if (A == B) and (B == C):
    print("O triângulo é equilátero!")
  elif (A == B) or (A == C) or (B == C):
    print("O triângulo é isósceles!")
  else:
    print("O triângulo é escaleno!")
else:
  print("Ao menos um dos valores indicados não servem para formar um triangulo!")
```
