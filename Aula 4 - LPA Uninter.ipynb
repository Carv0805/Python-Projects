##Contadores

```python
inicial = int(input('Qual valor deseja iniciar a contagem?'))
final = int(input('Qual valor deseja encerrar a contagem?'))

x = inicial
while (x <= final):
  #Verifica se o número é par
  if (x % 2 == 0):
    print(x)
  x = x + 1
```

##Acumuladores

```python
soma = 0
cont = 1
while (cont <= 5):
  x = float(input(f'Digite a {cont}ª nota:'))
  soma = soma + x
  cont = cont + 1
media = soma / 5
print(f'Média final: {media}')
```

##Interrompendo um loop com break

```python
print("Digite uma mensagem que irei repetir para você!")
print("Para encerrar escreva 'sair'.")
while True:
  texto = input('')
  print(texto)
  if texto == 'sair':
    break
print('Encerrando o programa...')
```

##Interrompendo um loop com continue

```python
while True:
  nome = input("Qual o seu nome?")
  if nome != "Vitor":
    continue # volta para o inicio do laço
  senha = input("Qual a sua senha?")
  if senha == "123456":
    break # encerra o laço
print("Acesso permitido")
```

#Estrutura de Repetição for

Assim como o while, essa estrutura repete um bloco de instruções **enquanto** uma condição se mantiver verdadeira.

No entanto, diferentemente do *while*, o *for* é empregado em situações e que o número de vezes que o laço irá executar é finito.

image.png

```python
for i in range(6):
  print(i)
```

```python
for i in range(1,6,1):
  print(i)
```

```python
for i in range(10,0,-2):
  print(i)
```

##Varredura de string

Podemnos passar por todos os caracteres de uma string usando um laço *for*

```python
frase = "Lógica de Programação e Algoritmos"
for i in range(0,len(frase),1):
  print(frase[i],end="")
```

##Exercício


Escreva um algoritmo que calcule a média dos números pares de 1 até 100 (1 e 100 inclusos).

```python
soma = 0
qtd = 0
for i in range(100,0,-2):
  if (i % 2 == 0):
    soma += i
    qtd += 1
media = soma / qtd
print(f'Média dos números pares de 1 a 100: {media}')
```

##Exercício
Escreva um algoritmo em Python que calcule a tabuada de todos os números de 1 até 10, e, para cada número, vamo calcular a tavuada mutiplicando-o pelo intervalo de 1 até 10.

```python
#2x while
tabuada = 0
while (tabuada <= 10):
  print(" ")
  print(f'Tabuada do {tabuada}:')
  i = 1
  while i <= 10:
    print(f'{tabuada} x {i} = {tabuada * i}')
    i += 1
  tabuada += 1
```

```python
#2x for
for tabuada in range(1,11):
  print(" ")
  print(f'Tabuada do {tabuada}:')
  for i in range(1,11):
    print(f'{tabuada} x {i} = {tabuada * i}')
```

```python
# while + for
tabuada = 1
while (tabuada <= 10):
  print(" ")
  print(f'Tabuada do {tabuada}:')
  for i in range(1,11):
    print(f'{tabuada} x {i} = {tabuada * i}')
  tabuada += 1
```

Exercício 2

Escreva um algoritmo que mostra, na tela, quatro produtos a serem comprados em uma lanchonete:


*   Coxinha - RS 5,00
*   Pastel - RS 7,00
*   Café - RS 4,00
*   Suco - RS 6,00

Faça um algoritmo em que voce seleciona o produto que quer comprar e a quantidade. Faça isso até que decida encerrar o programa.

Ao final, mostre o valor final do pedido a ser pago.




```python
print("LANCHONETE!")
print("1 - Coxinha - R$ 5,00")
print("2 - Pastel - R$ 7,00")
print("3 - Café - R$ 4,00")
print("4 - Suco - R$ 6,00")
print("5 - SAIR")

total = 0
while True:
  compra = int(input("Digite o número da opção desejada: "))
  if (compra == 1):
    qtd = int(input("Digite a quantidade desejada: "))
    total += (5 * qtd)
  elif (compra == 2):
    qtd = int(input("Digite a quantidade desejada: "))
    total += (7 * qtd)
  elif (compra == 3):
    qtd = int(input("Digite a quantidade desejada: "))
    total += (4 * qtd)
  elif (compra == 4):
    qtd = int(input("Digite a quantidade desejada: "))
    total += (6 * qtd)
  elif (compra == 5):
    break
  else:
    print("Produto inválida! Selecione outro.")

print(f'O total a ser gasto neste pedido é de R$ {total}')
```

Exercício 3

Escreva um algoritmo que leia um valor e que imprima a quantidade de cédulas necessárias para pagar esse mesmo valor.

Para simplificar, vamos trabalhar apenas com valores inteiros e com cédulas de RS100, RS50, RS20, RS10, RS5 e RS1.

```python
valor = int(input("Digite o valor em R$: "))

while True:
  if valor >= 100:
    cont100 = valor // 100
    valor -= cont100 * 100
    print(f"Cédulas de 100: {cont100}")
    if not valor:
      break

  elif valor >= 50:
    cont50 = valor // 50
    valor -= cont50 * 50
    print(f"Cédulas de 50: {cont50}")
    if not valor:
      break

  elif valor >= 20:
    cont20 = valor // 20
    valor -= cont20 * 20
    print(f"Cédulas de 20: {cont20}")
    if not valor:
      break

  elif valor >= 10:
    cont10 = valor // 10
    valor -= cont10 * 10
    print(f"Cédulas de 10: {cont10}")
    if not valor:
      break

  elif valor >= 5:
    cont5 = valor // 5
    valor -= cont5 * 5
    print(f"Cédulas de 5: {cont5}")
    if not valor:
      break

  else:
    cont1 = valor
    print(f"Cédulas de 1: {cont1}")
    break
```

Exercício 3

Um cinema cobra preços diferentes para ingressos, de acordo com a idade da pessoa. Se a pessoa tiver menos de 3 anos de idade, o ingresso será gratuito; se tiver entre 3 e 12 anos, o ingresso custará RS 15; se tiver mais de 12 anos, custará RS 30.

Escreva um laço em que voce pergunte a idade aos usuários e, então, informe-lhes o preço do ingresso do cinema. Encerre o laço, usando um *break* quando o usuário digitar *zero*

Após encerrar o laço, apresente na tela o total de pessoas que compraram ingressos, o total de dinheiro arrecadado e a média de idade das pessoas.


```python
pessoas = 0
ingresso = 0
acc_idades = 0
while True:
  idade = int(input("Digite sua idade: "))
  if (idade <= 3 and idade != 0):
    print("Ingresso gratuito!")
  elif (idade > 3 and idade <= 12):
    print("Ingresso custa R$ 15,00")
    ingresso += 15
  elif (idade > 12):
    print("Ingresso custa R$ 30,00")
    ingresso += 30
  else: #(idade == 0):
    break
  pessoas += 1
  acc_idades += idade
media= acc_idades / pessoas
print(f'Total de pessoas: {pessoas}')
print(f'Média de idades: {media}')
print(f'Total de dinheiro arrecadado: R$ {ingresso}')
```
